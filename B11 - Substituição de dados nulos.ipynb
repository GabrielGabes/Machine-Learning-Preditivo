# Criando dataframe sintetico desbalanceado

```python
from sklearn.datasets import make_classification

# Gerar dados sintéticos desbalanceados
x, y = make_classification(n_samples=7346, 
                           n_features=10, # qtd total de features
                           n_informative=7, # features realmente uteis para o modelo
                           n_redundant=3, # features que diz praticamente oque uma outra já fim
                           n_classes=2, 
                           n_clusters_per_class=3,  # Subgrupos
                           weights=[0.95, 0.05], 
                           class_sep=0.1,
                           random_state=3141592)

# Criar um DataFrame
df = pd.DataFrame(x, columns=[f'feature_{i}' for i in range(x.shape[1])])
df['target'] = y

# Ver a distribuição de classes
print(df['target'].value_counts(normalize=True))
print(df['target'].value_counts())

y = df['target']
df = df.drop('target', axis=1)

display(pd.concat([df.head(5), df.tail(5)]).T)
df
```

```python
import random
lista_numerica = sorted(random.sample(range(1, 7346), 1000))
lista_strings = [f"feature_{i}" for i in range(0, 10)]
print(lista_strings)

dicionario = {numero: random.choice(lista_strings) for numero in lista_numerica}
```

```python
df_bck = df.copy()

for chave in dicionario.keys():
    df[ dicionario[chave] ][chave] = np.nan
df.loc[list(dicionario.keys())[:10]].round(1).T
```

```python
from sklearn.impute import KNNImputer

# Imputação kNN
imputer = KNNImputer(n_neighbors=5)#, weights='distance', metric='nan_euclidean')
data_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
data_imputed.loc[list(dicionario.keys())[:10]].round(1).T
```

```python
df_comp = pd.DataFrame(columns=['coluna','original'])

for chave in dicionario.keys():
    coluna = dicionario[chave]
    original = df_bck[ dicionario[chave] ][chave]
    df_comp.loc[len(df_comp)+1] = [coluna, original]

df_comp_bck = df_comp.copy()
df_comp
```

```python
df_comp = df_comp_bck.copy()
from sklearn.model_selection import ParameterGrid

# Definindo os hiperparâmetros a serem testados
param_grid = {
    'n_neighbors': np.arange(3, 31, 3),
    'weights': ['distance','uniform'],
    'metric': ['nan_euclidean','manhattan','cosine'],
}

# Gerando todas as combinações de hiperparâmetros
grid = list(ParameterGrid(param_grid))
print('qtd total =', len(grid))
#grid[:5]
###############################################################################3

for params in grid:
    clf = KNNImputer(
        n_neighbors = params['n_neighbors'], 
        weights=params['weights'], 
        metric=params['metric']
        )
    try:
        data_imputed = pd.DataFrame(clf.fit_transform(df), columns=df.columns)
        
        #
        nome_coluna = 'm' + '_' + str(params['n_neighbors'])+ '_' + str(params['weights'])+ '_' + str(params['metric'])
        df_comp[nome_coluna] = np.nan
        
        count = 0
        for chave in dicionario.keys():
            coluna = dicionario[chave]

            original = df_bck[ dicionario[chave] ][chave]
            previsto = data_imputed[ dicionario[chave] ][chave]
            diff = abs(original - previsto)

            df_comp[nome_coluna][count] = diff
            count += 1
    except:
        pass
df_comp = df_comp.drop(['coluna','original'], axis=1)
print(df_comp.shape)
```

```python
df_long = df_comp.melt(var_name="Variável", value_name="Valor")
df_long['Valor'] = df_long['Valor']*100

import plotly.express as px
fig = px.violin(df_long, x="Variável", y="Valor", box=True)
fig.show()
```

```python
desc_stats = df_comp.describe().T*100
curtose_assimetria = pd.DataFrame({
    'curtose': df_comp.kurtosis(),
    'assimetria': df_comp.skew()
})
combined_df = pd.concat([desc_stats, curtose_assimetria], axis=1)

# Normalizando as colunas de curtose, assimetria, Q1 (25%), Q2 (50%) e Q3 (75%) para valores entre 0 e 1
Q1_norm = (combined_df['25%'] - combined_df['25%'].min()) / (combined_df['25%'].max() - combined_df['25%'].min())
Q2_norm = (combined_df['50%'] - combined_df['50%'].min()) / (combined_df['50%'].max() - combined_df['50%'].min())
Q3_norm = (combined_df['75%'] - combined_df['75%'].min()) / (combined_df['75%'].max() - combined_df['75%'].min())
curtose_norm = (combined_df['curtose'] - combined_df['curtose'].min()) / (combined_df['curtose'].max() - combined_df['curtose'].min())
assimetria_norm = (combined_df['assimetria'] - combined_df['assimetria'].min()) / (combined_df['assimetria'].max() - combined_df['assimetria'].min())

# Calculando o índice composto com peso igual para todas as métricas (valores menores priorizados)
# combined_df['indice_composto'] = (
#     0.2*(1 - curtose_norm) + 0.2*(1 - assimetria_norm) + 0.2*(1 - Q1_norm) + 0.2*(1 - Q2_norm) + 0.2*(1 - Q3_norm)
# ) / 5
combined_df['indice_composto'] = (
    (1 - curtose_norm) + (1 - assimetria_norm) + (1 - Q1_norm) + (1 - Q2_norm) + (1 - Q3_norm)
) / 5
combined_df['indice_composto2'] = (
    0.125 * (1 - curtose_norm) +
    0.125 * (1 - assimetria_norm) +
    0.25 * (1 - Q1_norm) +
    0.25 * (1 - Q2_norm) +
    0.25 * (1 - Q3_norm)
)

combined_df = combined_df.sort_values(by='indice_composto', ascending=False)
combined_df.round(2)
```
